```python
# liner regrassion
from sklearn.model_selection import train_test_split
```

```python
#convert string date to date_time
import datetime as dt
India_cases["date"] = pd.to_datetime(India_cases["date"])
India_cases.head()
```

```python
#converting date_time to ordinal
India_cases["date"]=India_cases["date"].map(dt.datetime.toordinal)
India_cases.head()
```

```python
x=India_cases["date"]
y=India_cases["total_cases"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
import numpy as np
```

```python
lr.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
India_cases.tail()
```

```python
y_pred = lr.predict(np.array(x_test).reshape(-1,1))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(x_test,y_pred)
```
